#### Code (15 marks)

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
```

### -- Reading the data:

```python
df = pd.read_csv('https://github.com/Eduardostca/ML_/blob/main/veh18.csv?raw=true')
df
```

```python
df.info()
```

### -- Isolating all numeric features:

```python
df_num = df.select_dtypes(include = 'number')
df_num
```

### -- Verifying missing values:

```python
df.isna().sum()
```

### -- Creating a baseline model:

```python
X = df_num.drop(['Comb Unadj FE - Conventional Fuel'], axis=1)
y = df_num['Comb Unadj FE - Conventional Fuel']
```

```python
oob_scores = []
predictions_1 =[]
for i in range(10):
    model_1 = RandomForestRegressor(n_estimators=100, n_jobs=-1, oob_score=True)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model_1.fit(X_train, y_train)
    oob_scores.append(model_1.oob_score_)

avg_oob_scores_1 = sum(oob_scores) / len(oob_scores)
print(f"Average OOB score over 10 runs for model 1: {avg_oob_scores_1:.4f}")
```
